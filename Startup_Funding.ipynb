```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df= pd.read_csv("startup_funding.csv")
```

```python
df
```

```python
print("Shape of the dataset:",df.shape)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
pd.set_option('display.float_format', '{:,.2f}'.format)
df['Amount in USD'] = (
    df['Amount in USD']
    .astype(str)                         
    .str.replace(',', '', regex=False)   
    .pipe(pd.to_numeric, errors='coerce') 
)
```

```python
df['Amount in USD'].fillna(df['Amount in USD'].mean(), inplace=True)
```

```python
df.dtypes
```

```python
df['Industry Vertical'].fillna(df['Industry Vertical'].mode()[0], inplace=True)
```

```python
df['SubVertical'].fillna(df['SubVertical'].mode()[0], inplace=True)
```

```python
df['City  Location'].fillna(df['City  Location'].mode()[0], inplace=True)
```

```python
df['Investors Name'].fillna(df['Investors Name'].mode()[0], inplace=True)
```

```python
df['InvestmentnType'].fillna(df['InvestmentnType'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date dd/mm/yyyy'], errors='coerce', format='%d/%m/%Y')
```

```python
df['Year'] = df['Date'].dt.year
```

## Number of Startups Founded in Each Year

```python
startups_per_year = df.groupby('Year')['Startup Name'].nunique()

plt.figure(figsize=(6, 5))
startups_per_year.plot(kind='bar', color='skyblue')
plt.title('Number of Startups Founded Each Year')
plt.xlabel('Year')
plt.ylabel('Number of Startups')
plt.xticks(rotation=45)
plt.show()
```

## Funding Amount Over Time

```python
funding_over_time = df.groupby('Year')['Amount in USD'].sum()

plt.figure(figsize=(8, 4))
funding_over_time.plot(kind='line', marker='o', color='green')
plt.title('Funding Amount Over Time')
plt.xlabel('Year')
plt.ylabel('Total Funding (USD)')
plt.grid()
plt.show()
```

## Most Funded Sectors

```python
most_funded_sectors = df.groupby('Industry Vertical')['Amount in USD'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
most_funded_sectors.plot(kind='bar', color='orange')
plt.title('Top 10 Most Funded Sectors')
plt.xlabel('Industry Vertical')
plt.ylabel('Total Funding (USD)')
plt.xticks(rotation=45)
plt.show()
```

## Most Funded Regions

```python
most_funded_regions = df.groupby('City  Location')['Amount in USD'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
most_funded_regions.plot(kind='bar', color='purple')
plt.title('Top 10 Most Funded Regions')
plt.xlabel('City')
plt.ylabel('Total Funding (USD)')
plt.xticks(rotation=45)
plt.show()
```

## Most Active Investors

```python
most_active_investors = df['Investors Name'].value_counts().head(10)

plt.figure(figsize=(10, 6))
most_active_investors.plot(kind='bar', color='teal')
plt.title('Top 10 Most Active Investors')
plt.xlabel('Investor Name')
plt.ylabel('Number of Investments')
plt.xticks(rotation=45)
plt.show()
```

## Heatmap of Investment Types vs. Sectors

```python
pivot_table = df.pivot_table(values='Amount in USD', index='InvestmentnType', columns='Industry Vertical', aggfunc='sum')

plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, annot=False, cmap='coolwarm', cbar=True)
plt.title('Heatmap of Investment Types vs. Sectors')
plt.xlabel('Industry Vertical')
plt.ylabel('Investment Type')
plt.show()
```

## Industries Most Funded in India

```python
indian_startups = df[df['City  Location'].str.contains('India', case=False, na=False)]

industry_funding_india = indian_startups.groupby('Industry Vertical')['Amount in USD'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
industry_funding_india.head(10).plot(kind='bar')
plt.title('Most Funded Industries in India')
plt.xlabel('Industry')
plt.ylabel('Total Funding (in USD)')
plt.xticks(rotation=45)
plt.show()
```
